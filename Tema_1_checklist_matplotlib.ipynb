## Creación de Gráficos Básicos

Descripción: Generar gráficos de líneas, barras, dispersión e histogramas.


```python
import matplotlib.pyplot as plt
import numpy as np

# Gráfico de líneas
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Gráfico de Líneas')
plt.xlabel('X')
plt.ylabel('sin(X)')
plt.show()
```

```python
# Gráfico de barras
categorias = ['A', 'B', 'C']
valores = [10, 20, 15]
plt.bar(categorias, valores)
plt.title('Gráfico de Barras')
plt.show()
```

```python
# Gráfico de dispersión
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
plt.title('Gráfico de Dispersión')
plt.show()
```

```python
# Histograma
datos = np.random.randn(1000)
plt.hist(datos, bins=30)
plt.title('Histograma')
plt.show()
```

## Personalización de Gráficos

Descripción: Añadir títulos, etiquetas, leyendas y personalizar colores y estilos.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='Sin(x)', color='blue', linestyle='--')
plt.plot(x, y2, label='Cos(x)', color='red')
plt.title('Funciones Trigonométricas')
plt.xlabel('X')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.show()
```

## Subgráficos (Subplots)

Descripción: Crear múltiples gráficos en una sola figura.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(2, 1, figsize=(8, 6))

axs[0].plot(x, y1, 'g-')
axs[0].set_title('Seno')

axs[1].plot(x, y2, 'b-')
axs[1].set_title('Coseno')

plt.tight_layout()
plt.show()
```

## Personalización de Ejes

Descripción: Configurar límites, escalas (logarítmica), y ticks en los ejes.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1, 100, 100)
y = np.log(x)

plt.plot(x, y)
plt.title('Escala Logarítmica en el Eje Y')
plt.xlabel('X')
plt.ylabel('log(X)')

plt.yscale('log')  # Escala logarítmica en Y
plt.xlim(0, 110)
plt.ylim(0.1, 5)
plt.show()
```

## Guardar Figuras

Descripción: Exportar gráficos a diferentes formatos (PNG, PDF, etc.) con control sobre la resolución y tamaño.

## Integración con Pandas para Visualización

Descripción: Utilizar las capacidades de graficado de pandas que se basan en matplotlib.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear un DataFrame de ejemplo
data = {
    'Año': [2018, 2019, 2020, 2021, 2022],
    'Ventas': [250, 300, 350, 400, 450],
    'Beneficios': [50, 60, 70, 80, 90]
}
df = pd.DataFrame(data)

# Graficar usando pandas
df.plot(x='Año', y=['Ventas', 'Beneficios'], kind='bar', figsize=(8,6))
plt.title('Ventas y Beneficios por Año')
plt.xlabel('Año')
plt.ylabel('Monto')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Crear un DataFrame de ejemplo
np.random.seed(0)
dates = pd.date_range('20230101', periods=100)
df = pd.DataFrame({
    'Fecha': dates,
    'Ventas': np.random.randint(100, 200, size=100),
    'Clientes': np.random.randint(50, 150, size=100),
    'Región': np.random.choice(['Norte', 'Sur', 'Este', 'Oeste'], size=100)
})
df.set_index('Fecha', inplace=True)

# Crear subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Gráfico de líneas para Ventas
df['Ventas'].plot(ax=axs[0, 0], color='blue', title='Ventas Diarias')
axs[0, 0].set_xlabel('Fecha')
axs[0, 0].set_ylabel('Ventas')

# Gráfico de barras para Clientes
df['Clientes'].plot(kind='bar', ax=axs[0, 1], color='green', title='Clientes Diarios')
axs[0, 1].set_xlabel('Fecha')
axs[0, 1].set_ylabel('Clientes')
# Reducir el número de ticks para mejorar la legibilidad
axs[0, 1].set_xticks(range(0, 100, 10))
axs[0, 1].set_xticklabels(df.index[::10], rotation=45)

# Gráfico de dispersión Ventas vs Clientes
df.plot.scatter(x='Ventas', y='Clientes', ax=axs[1, 0], color='red', title='Ventas vs Clientes')
axs[1, 0].set_xlabel('Ventas')
axs[1, 0].set_ylabel('Clientes')

# Gráfico de pastel para distribución por Región
región_counts = df['Región'].value_counts()
axs[1, 1].pie(región_counts, labels=región_counts.index, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Distribución de Ventas por Región')
axs[1, 1].axis('equal')  # Para que el pastel sea circular

plt.tight_layout()
plt.show()
```

## Gráficos Avanzados (Dual Axes, 3D)

Descripción: Crear gráficos con ejes duales y gráficos tridimensionales.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.exp(x)

fig, ax1 = plt.subplots()

ax1.set_xlabel('X')
ax1.set_ylabel('Sin(X)', color='tab:blue')
ax1.plot(x, y1, color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()  # Instanciar un segundo eje que comparte el mismo eje X
ax2.set_ylabel('Exp(X)', color='tab:red')
ax2.plot(x, y2, color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

plt.title('Gráfico con Ejes Duales')
fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Superficie 3D')
plt.show()
```
